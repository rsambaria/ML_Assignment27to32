In this assignment, students will be using the K-nearest neighbors
algorithm to predict how many points NBA players scored in the 2013-2014
season.

```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import Imputer
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import mean_squared_error 
from math import sqrt
```

```python
a=pd.read_csv(r'Downloads\nba_2013.csv')
a.sample(frac=1).head()
```

```python
a=a.drop(['player','season','season_end'],axis=1)
a.shape
```

```python
a.head()
```

```python
le = LabelEncoder()
a['pos'] = le.fit_transform(a['pos'])
a['bref_team_id'] = le.fit_transform(a['bref_team_id'])
```

```python
a.info()
```

```python
X=a.drop(['pts'],axis=1)
Y=a['pts']
Y.shape
```

```python
np.unique(X)
```

```python
imp = Imputer(missing_values="NaN", strategy='median', axis=1)
X = imp.fit_transform(X)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size = 0.3, random_state = 100)
np.unique(X_test)
```

```python
for K in range(25):
    K_value = K+1
    neigh = KNeighborsRegressor(n_neighbors = K_value, weights='uniform', algorithm='auto')
    neigh.fit(X_train, y_train) 
    y_pred = neigh.predict(X_test)
    error = sqrt(mean_squared_error(y_test,y_pred))
    print(error)
```

```python
neigh = KNeighborsRegressor(n_neighbors = 7, weights='uniform', algorithm='auto')
neigh.fit(X_train, y_train) 
y_pred = neigh.predict(X_test)
```

```python
pd.DataFrame({'Actual values':y_test,'Predicted values':y_pred})            
```
